```python
import numpy as np
import pandas as pd
from bokeh.layouts import row, widgetbox
from bokeh.models import HoverTool,CustomJS, Slider, ColumnDataSource
from bokeh.plotting import figure, output_notebook,output_file, show
import matplotlib.pyplot as plt
```

```python
output_notebook()
```

```python
data=pd.read_csv('indicator who bad_teeth.csv') 
```

```python
a=data["2004"]
b=np.array(a)
c=b.reshape(1,191)
x=c.ravel()
y=data.index.to_series()
coun=data["country"]
lcoun=list(coun)
# l=data["country"].values
# l
# print(x,lcoun)
source = ColumnDataSource(data=dict(x=x, y=y,name=lcoun))
data.plot.bar(y,title='country', figsize=(15,10), fontsize=12)
plot = figure(plot_width=500, plot_height=800,x_range=(0,5),title="indicator who bad_teeth",x_axis_label="bad teeth",y_axis_label="area name")
plot.circle('x','y', size=10, color="firebrick",fill_alpha=0.5,source=source)
plot.add_tools(HoverTool(tooltips=[("bad teeth number", "@x"),
    ("area name", "@name")]))
callback = CustomJS(args=dict(source=source), code="""
    var data = source.data;
    var B = min.value;
    var C = max.value;
    x = data['x']
    y = data['y']
    for (i = 0; i < x.length; i++) {
        if((x[i]-B)>=0 & (x[i]-C)<=0){ y[i]=i;x[i]=x[i];}
        else {x[i]=x[i];y[i]=NaN;}
    }
    source.change.emit();
""")
min_slider = Slider(title="min_bad_teeth released", start=0.1, end=5, value=0, step=0.1,callback=callback)
callback.args["min"] = min_slider
max_slider = Slider(title="max_bad_teeth released", start=0.1, end=5, value=5, step=0.1,callback=callback)
callback.args["max"] = max_slider
layout = row(plot,widgetbox(min_slider,max_slider))
output_file("indicator who bad_teeth.csv.html", title="indicator who bad_teeth.csv.py example")
show(layout)
```
